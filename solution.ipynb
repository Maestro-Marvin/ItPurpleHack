```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tema = pd.read_csv("tema.csv", low_memory=False)                  # 0.8970
vitya = pd.read_csv("vitya.csv", low_memory=False)                # 0.8975
andrew = pd.read_csv("andrew.csv", low_memory=False)              # 0.8950
```

```python
weight_tema = 1
weight_vitya = 1
weight_andrew = 1

ss = weight_tema + weight_vitya + weight_andrew

weight_tema /= ss
weight_vitya /= ss
weight_andrew /= ss

valid = (tema * weight_tema) + (vitya * weight_vitya) + (andrew * weight_andrew)

valid['id'] = range(200000, 200000 + 100000)

valid.to_csv("submissions/final_submissiom.csv", index=False)

valid
```
